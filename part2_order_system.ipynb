## Task 2 — Cart Operations

```python
# Starting with an empty cart
cart = []
```

```python
# Function to add item

def add_to_cart(item_name, quantity):
    
    # Checking if the item exists in menu
    if item_name not in menu:
        print(item_name, "does not exist in menu")
        return
    
    # Checking if the item is marked as available
    if menu[item_name]["available"] == False:
        print(item_name, "is currently unavailable")
        return
    
    # If the item is already in the cart, increasing its quantity instead of adding a duplicate entry
    found = False
    
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] = item["quantity"] + quantity
            found = True
            break
    
    # If the item doesn't exist in the menu or is unavailable, print an appropriate message and do not modify the cart.
    if found == False:
        cart.append({
            "item": item_name,
            "quantity": quantity,
            "price": menu[item_name]["price"]
        })
```

```python
# Function to remove item

def remove_from_cart(item_name):
    
    found = False
    
    for item in cart:
        if item["item"] == item_name:
            cart.remove(item)
            found = True
            break
    
    if found == False:
        print(item_name, "not found in cart")
```

```python
# Function to print cart

def print_cart():
    print("\nCurrent Cart:")
    
    if len(cart) == 0:
        print("Cart is empty")
        return
    
    for item in cart:
        print(item["item"], "x", item["quantity"], "₹", item["price"])
```

## Task 4 — Daily Sales Log Analysis 

```python
print("\n--- Revenue Per Day ---")

for date in sales_log:
    
    total_revenue = 0
    
    for order in sales_log[date]:
        total_revenue = total_revenue + order["total"]
    
    print(date, "→ ₹", total_revenue)
```

```python
max_revenue = 0
best_day = ""

for date in sales_log:
    
    total_revenue = 0
    
    for order in sales_log[date]:
        total_revenue = total_revenue + order["total"]
    
    if total_revenue > max_revenue:
        max_revenue = total_revenue
        best_day = date

print("\nBest-selling day:", best_day, "→ ₹", max_revenue)
```

```python
item_count = {}

for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            
            if item in item_count:
                item_count[item] = item_count[item] + 1
            else:
                item_count[item] = 1

# Find max
max_count = 0
most_ordered = ""

for item in item_count:
    if item_count[item] > max_count:
        max_count = item_count[item]
        most_ordered = item

print("\nMost ordered item:", most_ordered, "(", max_count, "orders )")
```
